```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
df[["Job","Marital","Eduaction","Default","Housing","Loan","Contact","Month","Poutcome","Y"]]=df[["Job","Marital","Eduaction","Default","Housing","Loan","Contact","Month","Poutcome","Y"]].apply(LabelEncoder().fit_transform)
```

```python
df.dtypes
```

```python
x = df.iloc[:,:-1].values
y = df.iloc[:,-1].values
```

```python
x
```

```python
y.shape
```

```python
x.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size = 0.3,random_state=45)
```

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
x_train = ss.fit_transform(x_train)
x_test = ss.transform(x_test)
```

```python
x_train
```

```python
x_test
```

**KNN Algorithm**

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors= 5)
knn.fit(x_train,y_train)
```

```python
y_pred = knn.predict(x_test)
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import confusion_matrix , accuracy_score , classification_report
```

```python
cm = confusion_matrix(y_test,y_pred)
cm
```

```python
acs = accuracy_score(y_test,y_pred)*100
acs
```

```python
cr = classification_report(y_test,y_pred)
print(cr)
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error
mse=mean_squared_error(y_test,y_pred)*100
mae=mean_absolute_error(y_test,y_pred)*100
print('mse=',mse)
print('mae=',mae)
```

**SVM Algorithm**

**Decision Tree**

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier(criterion='entropy', random_state=0)
dt.fit(x_train, y_train)
```

```python
y_pred_4 = dt.predict(x_test)
y_pred_4
```

```python
y_test
```

```python
cm_4 = confusion_matrix(y_test,y_pred_4)
cm_4
```

```python
acs_4 = accuracy_score(y_test,y_pred_4)*100
acs_4
```

```python
cr_4 = classification_report(y_test,y_pred_4)
print(cr_4)
```
